```python
# Author: Dennis Huynh
# Date: 02/05/2022

# Load dependencies
import os
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
# List of csv files
csvs = [f.name for f in os.scandir("./CORR_DATA/") if f.name.endswith(".csv")]

# Remove hidden directories
csvs = [f for f in csvs if not f.startswith('.')]

# Append directory as prefix to strings in list
csvs = ['./CORR_DATA/' + f for f in csvs]

for c in csvs:
    
    # Read file
    #df = pd.read_csv(c, sep=" ") # To fix output of anova_results.Rmd
    df = pd.read_csv(c)
    
    # Sort by p-value
    df.sort_values(by="Pr(>|t|)", inplace=True)
    
    # Reset index
   # df = df.reset_index(drop=True) # To fix output of anova_results.Rmd
    
    # Display
    display(df)
    
    # Print number of rows with significant p-value
    print("The number of rows with strong correlation are:", len(df[df["Pr(>|t|)"] < 0.05].index))
    
    # Save to csv
    df.to_csv(c, index=False)
```
